```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load the dataset

(xtrain,ytrain),(xtest,ytest)= datasets.cifar10.load_data()
```

```python
xtrain =xtrain/255
xtest =xtest/255
#scaling
```

```python
# ANN

model = tf.keras.Sequential([
                          layers.Flatten(input_shape=(32,32,3)), # input layer
                          layers.Dense(420, name="hiddenlayer1", activation="relu"), #hidden layer 1
                          layers.Dense(120,  name="hiddenlayer2",activation="relu"), # hidden layer 2
                          #keras.layers.Dropout(0.2),
                          layers.Dense(10,  name="hiddenlayer3",activation="softmax") # output layer
] )

#step 2
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

#step 3
model.fit(xtrain, ytrain,epochs=2, batch_size=50, validation_split=0.15)
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
# Prepare teh CNN model --fitlers/ pooling / making the image ready for our ANN model

cnn=models.Sequential([
                      layers.Conv2D(filters=32,kernel_size=(3,3), activation="relu", input_shape=(32,32,3)),
                      layers.MaxPooling2D((2,2)),

                      layers.Conv2D(filters=64,kernel_size=(3,3), activation="relu"),
                      layers.MaxPooling2D((2,2)),
                      
                      layers.Flatten(), 
                      layers.Dense(77, name="hiddenlayer1", activation="relu"), #hidden layer 1
                          #keras.layers.Dropout(0.2),
                      layers.Dense(10,  name="outputlayer",activation="softmax") # output layer
] )

#step 2
cnn.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

#step 3
cnn.fit(xtrain, ytrain,epochs=2, batch_size=50, validation_split=0.15)
```

```python
cnn.evaluate(xtest,ytest)
```
